```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
dataset=pd.read_csv('E:\project\Day 02\Day 02\iris.csv').values
```

```python
print(dataset,dataset.shape)
```

```python
data=dataset[:,0:4]
target=dataset[:,4]
print(data.shape,target.shape)
```

```python
print(data)
print(target)
```

```python
from keras.utils import np_utils
```

```python
# categoricl encoding
```

```python
new_target=np_utils.to_categorical(target)
print(target[0:10])
```

```python
print(new_target[0:10])
```

```python
from sklearn.model_selection import train_test_split
trin_data,test_data,trin_target,test_target=train_test_split(data,new_target,test_size=0.1)
```

```python
print(trin_data,trin_data.shape)
```

```python
print(test_data,test_data.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model=Sequential() #empty sequantal type neural network
model.add(Dense(64,input_dim=4,activation='relu')) #add layer(1st hidden layer)
model.add(Dense(32,activation='relu')) #2nd hidden layer
model.add(Dense(3,activation='softmax'))#output layer

#Compile ths model
model.compile(loss='categorical_crossentropy',optimizer='adaGrad',metrics=['accuracy'])
```

```python
model.summary()
```

```python
for layer in model.layers:
    print('=============',layer.name,'==============')
    parameter=layer.get_weights()
    print(parameter)
```

```python
# training the model
model.fit(trin_data,trin_target,epochs=1000,validation_split=0.1)
```

```python
#done trin neural network
#cheak acuracy of neural network
```

```python
from matplotlib import pyplot as plt
plt.plot(model.history.history['loss'],label='loss')
plt.plot(model.history.history['val_loss'],label='val_loss')
plt.xlabel("#epoch")
plt.legend()
plt.ylabel('loss')
plt.show()
```

```python
plt.plot(model.history.history['accuracy'],label='accuracy')
plt.plot(model.history.history['val_accuracy'],label='val_accuracy')
plt.legend()
plt.xlabel("#epoch")
plt.ylabel('loss')
plt.show()
```

```python
#get prediction
import numpy as np
test_data=np.array([5.3,2.3,4.7,2.1]).reshape(1,-1)
predicted_traget=model.predict(test_data)
print(predicted_traget)
```
